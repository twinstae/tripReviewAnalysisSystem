```python
ratings = []
reviews = []

for data in soup.find_all("div", class_="location-review-review-list-parts-RatingLine__bubbles--GcJvM"):
  ratings.append(int(data.contents[0]['class'][1][-2:]) / 10)

for data in soup2.find_all("q", class_="location-review-review-list-parts-ExpandableReview__reviewText--gOmRC"):
  reviews.append(data.contents[0].get_text())
```

```python
print(ratings[:5])
```

```python
df = pd.DataFrame({'Rating' : ratings, 'Review' : reviews})
df.head()
```
